```python
import os 
import tarfile
import json
import datetime
# set the directory path of data download
dir_path = '/home/ajay/Documents/Sentinel_thumbnails/thumnbails'

dir_contents = os.listdir(dir_path)
date_format = '%Y-%m-%dT%H:%M:%SZ'
 
print(dir_contents)

for item in dir_contents:
    # RENAME SENTINEL1 FILES, ADD FUNCTIONALITY FOR SENTINEL2 AND LANDSAT FILES IN FUTURE
    if item == 'SENTINEL1':
        current_dir = os.path.join(dir_path, item)
        print( "current dir:", current_dir)
        for filename in os.listdir(current_dir):
            file_path = os.path.join(current_dir, filename)
            print("sentinel1 folders:", file_path)

            for filename1 in os.listdir(file_path):
                print(filename1)

                # check if the file is a .tar file
                if filename1.endswith('.tar'):
                    print("folder found")
                    # extract the tar file
                    with tarfile.open(os.path.join(file_path, filename1), 'r:') as tar:
                        tar.extractall(file_path)
                        tar.close()

                        with open(os.path.join(file_path, 'request.json')) as f:
                            data = json.load(f)
                            f.close()
                            # FIND TIMESTAMP FROM JSON FILE
                            timestamp = data['request']['payload']['input']['data'][0]['dataFilter']['timeRange']['from']
                            # FIND ORIENTIATION FROM JSON FILE
                            orientation = data['request']['payload']['input']['data'][0]['dataFilter']['orbitDirection'][0:3].lower()
                            dt = datetime.datetime.strptime((timestamp), date_format)

                            print (dt.year, dt.month, dt.day)

                            # RENAME THE FILES, COULD RENAME THE JSON FILE TOO SO THAT ALL FILES CAN BE SAVED IN A SINGLE FOLDER
                            os.rename(f'{file_path}/vv_band.tif', f'{file_path}/{dt.year}_{dt.month}_{dt.day}_vv_{orientation}.tif')
                            os.rename(f'{file_path}/vh_band.tif', f'{file_path}/{dt.year}_{dt.month}_{dt.day}_vh_{orientation}.tif')

```
